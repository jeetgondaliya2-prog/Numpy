```python
import numpy as np    # fancy indexing & indexing with boolean arrays 
```

```python
arr=np.arange(24).reshape(6,4)
```

```python
arr
```

```python
arr[[0,2,4]]           # 1,3 and 5 row
```

```python
arr1=np.random.randint(low=1,high=100,size=20).reshape(4,5)           # random array
arr1   
```

```python
arr1[0]                             # first row
```

```python
arr1[0][2]                      # first row third element
```

```python
 arr1[arr1>50]                        # element greater than 50   & indxeing using boolean arrays
```

```python
arr1[(arr1>50) & (arr1%2!=0)]
```
