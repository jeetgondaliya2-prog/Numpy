```python
import numpy as np                     #operations
```

```python
arr1=np.array([1,2,3,4,5,6])
```

```python
arr2=np.array([4,5,6,7,8,9])
```

```python
arr1-arr2                         # element  wise subtraction
```

```python
arr1*arr2                        # element wise multiplication
```

```python
arr1*3                          # every element will multiply with 3
```

```python
arr2>5                   # compare every element
```

```python
arr3=np.arange(6).reshape(2,3)
```

```python
arr4=np.arange(6,12).reshape(3,2)
```

```python
arr3
```

```python
arr4
```

```python
arr3.dot(arr4)             #dot product 
```

```python
arr4.max()               #maximum 
```

```python
arr4.min()               # minimum
```

```python
arr4.max(axis=0)          #every column maximum elemenet
```

```python
arr4.min(axis=1)            # every row minimum element
```

```python
arr4
```

```python
arr4.sum()           # sum of all elements
```

```python
arr4.sum(axis=0)     # every column sum
```

```python
arr4.mean()         # mean of all elements   also applicable for rows and column by using axis
```

```python
arr4.std()          # standard deviation     also applicable for rows and column by using axis
```

```python
np.sin(arr4)             #sin
```

```python
np.median(arr4)           # median
```

```python
np.exp(arr4)               # exponent  
```
