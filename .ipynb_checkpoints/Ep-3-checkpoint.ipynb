```python
import numpy as np           # python list  vs numpy array 
```

```python
list1=range(100              #list
list1  
```

```python
arr1 = np.arange(100)         #array
arr1
```

```python
import sys                  #for memory
```

```python
print(sys.getsizeof(87)*len(list1))         # memory occupy by the integer 87 (beacause 87 is in the range) * size of list
```

```python
print(arr1.itemsize * arr1.size)      #size of array is 100 and each integer is of 8 byte
```

```python
# numpy array store less memory than python list
```

```python
import time                    #for time
```

```python
x=range(100000)
y=range(100000,200000)
start_time=time.time()
c=[ x+y for x,y in zip(x,y)]
print(time.time()-start_time)
```

```python
a=np.arange(100000)
b=np.arange(100000,200000)
start_time=time.time()
c=a+b                      
print(time.time()-start_time)
```

```python
#numpy array is faster than python list
```
