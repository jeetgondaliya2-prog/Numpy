```python
arr=np.arange(6,12).reshape(3,2)
```

```python
arr1=np.arange(6).reshape(2,3)
```

```python
arr2=np.arange(12,18).reshape(2,3)
```

```python
arr1
```

```python
arr2
```

```python
np.hstack((arr1,arr2))              # horizontal stack
```

```python
 np.vstack((arr1,arr2))             # vertical stack
```

```python
np.hsplit(arr1,3)                 # horizontal split
```

```python
np.vsplit(arr1,2)                # vertical split
```
