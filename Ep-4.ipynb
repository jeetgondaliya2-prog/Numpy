```python
import numpy as np   #indexing,slicing and iteration 
```

```python
arr1=np.arange(50)
arr1
```

```python
arr1=np.arange(50).reshape(5,10)          #for reshaping
arr1
```

```python
arr1[2]                                  # for 3rd row
```

```python
arr1[:,2]                               # for 3rd column
```

```python
arr1[:,3:6]                            # for 4th to 6th column
```

```python
arr1[2:4,4:6]                        # for 24,25,34,35 
```

```python
for i in arr1:                     # iterate over row
    print(i)
```

```python
for i in np.nditer(arr1):       # iterate every elements 
    print(i)
```
